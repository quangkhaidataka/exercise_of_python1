Problem 1: Write a program which will find all such numbers which are symetric (such as 11, 121, 1221,...) between 10 and 2000 (both included). The numbers obtained should be printed in a comma- separated sequence on a single line.

```python
def is_symetric(num):
    num_list=list(str(num))
    reverse_list=num_list[::-1]
    if num_list== reverse_list:
        return True
    else:
        return False
```

```python
symetric_list=[]
for i in range(10,2001):
    if is_symetric(i):
        symetric_list.append(i)
print('All such numbers which are symetric between 10 and 2000 (both included) are: {}'.format(str(symetric_list).strip("[]")))
```

Problem 2: Write a program which will find all such numbers which include at least 3 duplicated digits (such as 111, 1222,...) between 100 and 2000 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.


```python
def is_3_duplicate(num):
    num_list=list(str(num))
    occurence_list=[num_list.count(i) for i in set(num_list)]
    greater_than3_list=[e>2 for e in occurence_list]
    if sum(greater_than3_list)> 0:
        return True
    else:
        return False
        
```

```python
duplicate_numbers=[]
for i in range(100,2001):
    if is_3_duplicate(i):
        duplicate_numbers.append(i)
```

```python
print('The at least 3 duplicated digits numbers are: {}'.format(str(duplicate_numbers).strip("[]")))
```

Problem 4: Write a program which will receive an integer larger than 0 and present that integer in binary format.


```python
def represent(number):
    new_list=[]
    while number >0:
        a=int(math.log2(number))
        new_list.append(a)
        number=number-2**a
    final_list=[10**t for t in new_list]
    print(sum(final_list))
```

```python
represent(1456)
represent(135)
```

Problem 6: Write a program which will find all primes between 1000 and 3000 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
def find_number_divisor(n):
    list_divisors=[n]
    for i in range(1,int(n/2)+1):
        if n%i==0:
            list_divisors.append(i)
    return len(list_divisors)
```

```python
prime_list=[]
for i in range(1000,3001):
    if find_number_divisor(i)==2:
        prime_list.append(i)
```

```python
print('All primes between 1000 and 3000 (both included) are {}'.format(str(prime_list).strip("[]")))
```

Problem 7: Write a program which will receive an integer larger than 0 and respond all perfect numbers smaller than that integer. A perfect number is sum of its divisors except itself, such as 6 = 1+2+3. The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
def find_divisor(n):
    list_divisors=[n]
    for i in range(1,int(n/2)+1):
        if n%i==0:
            list_divisors.append(i)
    return list_divisors
```

```python
def is_perfect(num):
    if sum(find_divisor(num))==2*num:
        return True
    else:
        return False
```

```python
def search_perfect_number(num):
    perfect_number_list=[]
    for i in range(1,num):
        if is_perfect(i):
            perfect_number_list.append(i)
    return perfect_number_list
            
```

```python
search_perfect_number(10000)
```

```python
print('All perfect numbers smaller than input integer are: {}'.format(str(search_perfect_number(10000)).strip("[]")))
```

Problem 8: With a given integral number n, write a program to generate a dictionary that contains (i, i*i*i) such that is an integral number between 1 and n (both included). and then the program should print the dictionary. 

```python
def create_dict(n):
    new_dict={}
    for i in range(1,n+1):
        new_dict[i]=i*i*i
    return new_dict
```

Problem 9: Write a program which accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 3 or not. The numbers that are divisible by 3 are to be printed in a comma separated sequence. Example: 0100,0011,1010,1101

```python
items = []
num = [x for x in input().split(',')]
for p in num:
    x = int(p, 2)
    if not x%3:
        items.append(p)
print(','.join(items))
```

Problem 10: Write a program which will receive an integer larger than 0 and respond if that integer is sum of
three distinct squared numbers, such as 38 = 2^2 + 3^2 + 5^2 .

```python
#Create the function to check whether that integer is sum of two distinct squared numbers

def is_sum_of_2_squared(num):
    new_list=[]
    for i in range(1,int(np.sqrt(num))+1):
        if (np.sqrt(num-i**2)==int(np.sqrt(num-i**2)) and np.sqrt(num-i**2)!=i and np.sqrt(num-i**2)>0):
            new_list.append(i)
    if len(new_list)>=1:
        return True
    else:
        return False  
```

```python
#Create the function to check whether that integer is sum of three distinct squared numbers
def is_sum_of_3_squared(num):
    new_list=[]
    for i in range(1,int(np.sqrt(num))+1):
        if (is_sum_of_2_squared(num-i**2)):
            new_list.append(i)
    if len(new_list)>=1:
        print('This number is sum of three distinct squared numbers')
    else:
        print('This number is not sum of three distinct squared numbers') 
```
